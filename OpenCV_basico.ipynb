# Leer la imagen utilizando la función imread. Para este ejemplo utilizaremos la imagen del mandril (Se puede descargar

```python
#img_raw = cv2.imread('imagen.jpg') 
img_raw = cv2.imread('//content/drive/MyDrive/img/mandalas.jpg') # Lee una imagen
```

# Ver la imagen

```python
plt.imshow(img_raw)
```

OpenCV y matplotlib tienen diferentes órdenes de colores primarios. Mientras que OpenCV lee imágenes en forma de BGR, matplotlib, por otro lado, sigue el orden RGB.

```python
img_rgb = cv2.cvtColor(img_raw, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

# Guardar imágenes

```python
cv2.imwrite('//content/drive/MyDrive/img/final_imagen.jpg',img_rgb)
```

# Operaciones básicas con imágenes

Dibujar en imágenes

```python
image_blank = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(image_blank)
```

# **LINEA**

```python
#cv2.shape(line, rectangle etc)(image,Pt1,Pt2,color,thickness)
#Dibujar una diagonal roja con 5 px de ancho
img = np.zeros(shape=(512,512,3),dtype=np.int16)
line_red = cv2.line(img,(0,0),(512,512),(255,0,0),5)
plt.imshow(line_red)
```

```python
#cv2.shape(line, rectangle etc)(image,Pt1,Pt2,color,thickness)
#Dibujar una diagonal verde con 5 px de ancho
line_green = cv2.line(img,(0,0),(512,512),(0,255,0),5)
plt.imshow(line_green)
```

# **RECTÁNGULO**

```python

#Dibujar una rectangulo azul con 5 px de ancho
rectangle_1 = cv2.rectangle(img,(384,0),(510,128),(0,0,255),5)
plt.imshow(rectangle_1)
```

```python
#Dibujar una circulo  azul con 5 px de ancho
circle_1 = cv2.circle(img,(447,63),63,(0,0,255), -1) #-1 es el circlulo relleno
plt.imshow(circle_1)
```

# Escribir en **imágenes**

```python
font = cv2.FONT_HERSHEY_SIMPLEX
text = cv2.putText(img,'FIEE_UNI',(10,500),font,4,(88,6,21),2,cv2.LINE_AA)
plt.imshow(text)
```
